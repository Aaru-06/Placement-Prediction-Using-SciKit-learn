**Prediction using Decision Tree Classifier**

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,Y,test_size=0.3)
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
predicted_y=dtc.predict(x_test)
```

```python
print("Accuracy of the Decision Tree Classifier is %.1f%%"%(100*accuracy_score(y_test,predicted_y)))
```

**Prediction using Nearest Centroid Model**

```python
from sklearn.neighbors import NearestCentroid
```

```python
knc=NearestCentroid()
```

```python
y_train_knc=np.array(y_train).reshape(-1)
y_train_knc.shape
```

```python
knc.fit(x_train,y_train_knc)
```

```python
predicted_y_knc=knc.predict(x_test)
```

```python
print("Accuracy of the Nearest Centroid Classifier is %.1f%%"%(100*accuracy_score(np.array(y_test).reshape(-1),predicted_y_knc)))
```

**Prediction using KNN(K-Nearest Neighbors) Model**

**Prediction using Multi-layer Perceptron Neural Network**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
```

```python
ss=StandardScaler()
```

```python
ss.fit(x_train)
```

```python
x_train=ss.transform(x_train)
x_test=ss.transform(x_test)
```

```python
mlpnn=MLPClassifier(activation="tanh",max_iter=1000,hidden_layer_sizes=(13,13,13))
```

```python
mlpnn.fit(x_train,np.array(y_train).reshape(-1))
```

```python
predicted_y_mlpnn=mlpnn.predict(x_test)
```

```python
print("Accuracy of the Multilayer Perceptron Neural Network Classifier is %.1f%%"%(100*accuracy_score(np.array(y_test).reshape(-1),predicted_y_mlpnn)))
```
